## Correlação

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('dataset/camera.csv')

plt.figure(figsize=(30,20))

sns.heatmap(
    dataset.corr(),
    annot = False,
    linewidths = 2,
    center = 0
)

plt.savefig('imagens/corr.png')
```
